```python
import pandas as pd

reviews = pd.read_csv("../../input/wine-reviews/winemag-data_first150k.csv", index_col=0)
reviews.head(3)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotnine as pln

%matplotlib inline

sns.set_style('whitegrid')
```

## Subplotting

```python
import matplotlib.pyplot as plt
fig, axarr = plt.subplots(2, 1, figsize=(12, 8))
```

```python
axarr
```

```python
ig, axarr = plt.subplots(2, 1, figsize=(12, 10))

_ = reviews['points'].value_counts().sort_index().plot.bar(ax=axarr[0], fontsize=16)
axarr[0].set_title('points value counts', fontsize=20)

_ = reviews['province'].value_counts().head(20).plot.bar(ax=axarr[1], fontsize=16)
axarr[1].set_title('province top 20 value counts', fontsize=20)

ig.tight_layout(h_pad=0.5, w_pad=0)#调整整体空白
plt.subplots_adjust(wspace=0, hspace=0.5)#调整子图间距
_ = sns.despine(bottom=True, left=True)
```

```python
fig, axarr = plt.subplots(2, 2, figsize=(12, 8))
```

```python
fig, axarr = plt.subplots(2, 2, figsize=(12, 8))

_ = reviews['points'].value_counts().sort_index().plot.bar(
    ax=axarr[0][0]
)

_ = reviews['province'].value_counts().head(20).plot.bar(
    ax=axarr[1][1]
)
```

```python
fig, axarr = plt.subplots(2, 2, figsize=(12, 8))

reviews['points'].value_counts().sort_index().plot.bar(
    ax=axarr[0][0], fontsize=12, color='mediumvioletred'
)
axarr[0][0].set_title("Wine Scores", fontsize=18)

reviews['variety'].value_counts().head(20).plot.bar(
    ax=axarr[1][0], fontsize=12, color='mediumvioletred'
)
axarr[1][0].set_title("Wine Varieties", fontsize=18)

reviews['province'].value_counts().head(20).plot.bar(
    ax=axarr[1][1], fontsize=12, color='mediumvioletred'
)
axarr[1][1].set_title("Wine Origins", fontsize=18)

reviews['price'].value_counts().plot.hist(
    ax=axarr[0][1], fontsize=12, color='mediumvioletred'
)
axarr[0][1].set_title("Wine Prices", fontsize=18)

plt.subplots_adjust(hspace=.3)

import seaborn as sns
sns.despine()
```
