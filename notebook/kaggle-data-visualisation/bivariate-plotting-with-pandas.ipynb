## Exercises

```python
pokemon = pd.read_csv("../../input/pokemon/pokemon.csv", index_col=0)
pokemon.head()
```

```python
pokemon.columns
```

```python
_ = pokemon.plot.scatter(x='attack', y='defense')
```

```python
_ = pokemon.plot.hexbin(x='attack', y='defense', gridsize=15)
```

```python
pokemon_stats_legendary = pokemon.groupby(['is_legendary', 'generation']).mean()[['attack', 'defense']]
_ = pokemon_stats_legendary.plot.bar(stacked=True)
```

```python
metrics = ['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']
pokemon_stats_by_generation = pokemon.groupby('generation').mean()[metrics]
```

```python
_ = pokemon_stats_by_generation.plot.line()
```
