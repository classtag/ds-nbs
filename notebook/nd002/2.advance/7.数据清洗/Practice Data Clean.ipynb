### Missing values

````python
feature_missing_count = postings.isna().sum()
feature_missing_rate = feature_missing_count / len(postings)
feature_missing_rate_rank = feature_missing_rate.rank(method='dense', ascending=False)

df_feature_missing = pd.concat([feature_missing_count,feature_missing_rate, feature_missing_rate_rank], axis=1)
df_feature_missing.columns = ['count', 'rate', 'rank']
df_feature_missing.sort_values('rank')
````

````python
df_feature_missing[df_feature_missing['rate'] > 0.7].index.tolist() # select missing rate more than 70% features
````

````python
df_feature_missing[(df_feature_missing['count']>0) & (df_feature_missing['rate'] <= 0.7)].index.tolist()
````

#### Define
选择起始日期列中含有 "As soon as possible"、"Immediately"等词的所有记录，并将这些内容替换为 "ASAP"
